```
df = pd.read_csv("/content/drive/My Drive/colabs/rakuten/FIdoc/test.tsv",encoding="shift-jis",delimiter='\t')
```

```
print(df.columns)
df = df.loc[:,['FI','全請求項(公開公報)']]
print(df)
```

```
dn = len(df)
fis = {}
for i in range(dn):
  fi = df.FI.iat[i]
  fi = fi.replace(" ","")
  fis[i] = fi.split(",")
  

print(fis)
```

```
# "id -> fi" と "fi -> id" の辞書を作成
id2FI, FI2id = {},{}
for i in range(len(fis)):
    for j in fis[i]:
        if j not in FI2id:
            N = len(FI2id)
            FI2id[j] = N
            id2FI[N] = j
N_FI2id = len(FI2id) #FI数
print(N_FI2id)
```

```
id2FI_S = {}
for i in range(len(id2FI)):
    id2FI_S[i] = [id2FI[i][:1],id2FI[i][1:3],id2FI[i][3:4],re.split("[/:]",id2FI[i][4:])]
    id2FI_S[i][3][0] = format(int(id2FI_S[i][3][0]),'03x')
    if len(id2FI_S[i][3][1]) == 3:
        id2FI_S[i] = [id2FI_S[i][0],id2FI_S[i][1],id2FI_S[i][2],
                     id2FI_S[i][3][0],str(0)+str(0)+id2FI_S[i][3][1][:2],str(0)+str(0)+id2FI_S[i][3][1][2:]]
    elif len(id2FI_S[i][3][1]) == 4:
        id2FI_S[i] = [id2FI_S[i][0],id2FI_S[i][1],id2FI_S[i][2],
                     id2FI_S[i][3][0],id2FI_S[i][3][1],"000"]
    elif len(id2FI_S[i][3][1]) == 5:
        id2FI_S[i] = [id2FI_S[i][0],id2FI_S[i][1],id2FI_S[i][2],
                     id2FI_S[i][3][0],str(0)+str(0)+id2FI_S[i][3][1][:2],id2FI_S[i][3][1][2:]]
    elif len(id2FI_S[i][3][1]) == 6:
        id2FI_S[i] = [id2FI_S[i][0],id2FI_S[i][1],id2FI_S[i][2],
                     id2FI_S[i][3][0],str(0)+id2FI_S[i][3][1][:3],id2FI_S[i][3][1][3:]]
    elif len(id2FI_S[i][3][1]) == 7:
        id2FI_S[i] = [id2FI_S[i][0],id2FI_S[i][1],id2FI_S[i][2],
                     id2FI_S[i][3][0],id2FI_S[i][3][1][:4],id2FI_S[i][3][1][4:]]
    else:
        id2FI_S[i] = [id2FI_S[i][0],id2FI_S[i][1],id2FI_S[i][2],
                     id2FI_S[i][3][0],str(0)+str(0)+id2FI_S[i][3][1],"000"]
id2FI_S[0]
```

```
data_fi = []
for doc in fis:
  data_fi.append([])
  for fi in fis[doc]:
    id = FI2id[fi]
    out = id2FI_S[id]
    data_fi[-1].append(out)
print(len(data_fi))
```

```
docs = []
for i in range(dn):
  docs.append([])
  doc = df["全請求項(公開公報)"].iat[i]
  doc = re.split("【請求項\d】",doc)
  doc.pop(0)
  for i in doc:
    text = ''.join(i.split())
    text = text.replace("、","")
    text = text.replace("。","")
    text = re.sub("請求項\d+(((又は)|，|,|~|(または)|～|(乃至)|(ないし)|(から))?(請求項)?)*(\d+)?((までの)|の)?(うち)?((何れか)|(いずれか))?((一項)|(1項)|(ひとつ)|(一つ))?(に)?(記載の)?(記載された)?(もの)?(において)?","",text)
    text = re.sub("第\d+","",text)
    text = re.sub("前記[の]?", "",text)
    text = re.sub("前述[の]?", "",text)
    text = re.sub("\(.+?\)", "", text)
    text = re.sub("（.+?）", "", text)
    docs[-1].append(text)
print(docs)
```

```
wakati = MeCab.Tagger("-Owakati")
claim_wakati=[]
for p in docs:
  claim_wakati.append([])
  for text in p:
    result = wakati.parse(text)
    result = result.split()
    claim_wakati[-1].append(result)
print(len(claim_wakati))
```

```
data_all = []
for i in range(len(data_fi)):
  data_all.append([data_fi[i],claim_wakati[i]])
df_out = pd.DataFrame(data_all,columns=["FI","Text"])
df_out.to_csv("/content/drive/My Drive/colabs/rakuten/data_maked.csv")
```
